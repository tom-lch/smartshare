```python
# 判断第一次出现之后到现在为止最大涨幅和最大跌幅
from addons.stock_manage.models.stock_manage import StockLong, StockDaily
import pandas as pd
from peewee import *
import datetime

res = (StockLong.select().order_by(StockLong.id))
for raw in res:
    stock_day_data = (StockDaily.select(
                        StockDaily.ts_code, 
                        StockDaily.trade_date, 
                        StockDaily.open,
                        StockDaily.high, 
                        StockDaily.low, 
                        StockDaily.close, 
                        StockDaily.pre_close, 
                        StockDaily.pct_chg,
                        StockDaily.change,
                        StockDaily.vol, 
                        StockDaily.amount).where((StockDaily.ts_code==raw.ts_code) & (StockDaily.trade_date >= f"{raw.trade_date + datetime.timedelta(1)}")))
    if not stock_day_data:
        continue
    res_daily = pd.DataFrame(list(stock_day_data.dicts()))
    start_price = res_daily.iloc[0, 2]
    min_price = res_daily['low'].min()
    max_price = res_daily['high'].max()
    max_up = round((max_price / start_price - 1) * 100, 2)
    min_low = round((min_price / start_price - 1) * 100, 2)
    print(raw.ts_code, raw.trade_date, start_price, min_price, max_price, max_up, min_low)
    res_update = StockLong.update({
        "start_price": start_price,
        "max_price": max_price,
        "min_price": min_price,
        "max_up": max_up,
        "min_low": min_low,
    }).where(StockLong.id == raw.id).execute()
    print(res_update)
          
```
